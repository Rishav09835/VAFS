```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU
from tensorflow.keras.optimizers import Adam

# Paths (update accordingly)
data_folder = r"C:\Users\USER\Downloads\stockprediction\stockdata\stockdata" # <-- replace with your actual path
output_folder = r"C:\Users\USER\Downloads\VAFS-main\VAFS-main\gru predictions"
os.makedirs(output_folder, exist_ok=True)

# Get all CSV files
company_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

# Loop over each stock file
for filename in company_files:
    try:
        print(f"\nüìÇ Training on: {filename}")

        # Load and prepare data
        df = pd.read_csv(os.path.join(data_folder, filename))
        df['Date'] = pd.to_datetime(df['Date'])
        df.sort_values('Date', inplace=True)

        full_df = df.copy()  # To save predictions later

        # Use only 'Close' for GRU
        close_data = df['Close'].values.reshape(-1, 1)
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(close_data)

        # Create sequences
        sequence_length = 60
        x, y = [], []
        for i in range(sequence_length, len(scaled_data)):
            x.append(scaled_data[i-sequence_length:i])
            y.append(scaled_data[i])
        x, y = np.array(x), np.array(y)

        # Train-test split
        split = int(len(x) * 0.8)
        x_train, x_test = x[:split], x[split:]
        y_train, y_test = y[:split], y[split:]

        # Build GRU model
        model = Sequential()
        model.add(GRU(50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
        model.add(GRU(50))
        model.add(Dense(1))
        model.compile(optimizer=Adam(), loss='mean_squared_error')
        model.fit(x_train, y_train, epochs=10, batch_size=1, verbose=0)

        # Predict
        predictions = model.predict(x_test)
        predictions = scaler.inverse_transform(predictions)

        # Add prediction column to full DataFrame
        full_df['Prediction'] = np.nan
        full_df.iloc[-len(predictions):, full_df.columns.get_loc('Prediction')] = predictions.flatten()

        # Save result to output folder
        output_path = os.path.join(output_folder, filename.replace('.csv', '_with_prediction.csv'))
        full_df.to_csv(output_path, index=False)
        print(f"‚úÖ Saved predictions to: {output_path}")

        # Plot results
        plt.figure(figsize=(12, 6))
        plt.plot(df['Date'], df['Close'], label='Actual')
        plt.plot(df.iloc[-len(predictions):]['Date'], predictions, label='Predicted')
        plt.title(f'{filename} - Stock Price Prediction (GRU)')
        plt.xlabel('Date')
        plt.ylabel('Stock Price')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"‚ùå Error processing {filename}: {e}")
```
