```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU
from tensorflow.keras.optimizers import Adam

# Paths
data_folder = r"C:\Users\USER\Downloads\stockprediction\stockdata\stockdata"
output_folder = r"C:\Users\USER\Downloads\VAFS-main\VAFS-main\gru predictions"
os.makedirs(output_folder, exist_ok=True)

# Get all CSV files
company_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

# Lists to store metrics for all stocks
mse_list = []
rmse_list = []
mae_list = []
r2_list = []
accuracy_list = []

# Loop over each stock file
for filename in company_files:
    try:
        print(f"\nüìÇ Training on: {filename}")

        # Load and prepare data
        df = pd.read_csv(os.path.join(data_folder, filename))
        df['Date'] = pd.to_datetime(df['Date'])
        df.sort_values('Date', inplace=True)

        full_df = df.copy()  # To optionally save predictions

        # Use only 'Close' for GRU
        close_data = df['Close'].values.reshape(-1, 1)
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(close_data)

        # Create sequences
        sequence_length = 60
        x, y = [], []
        for i in range(sequence_length, len(scaled_data)):
            x.append(scaled_data[i-sequence_length:i])
            y.append(scaled_data[i])
        x, y = np.array(x), np.array(y)

        # Train-test split
        split = int(len(x) * 0.8)
        x_train, x_test = x[:split], x[split:]
        y_train, y_test = y[:split], y[split:]

        # Build GRU model
        model = Sequential()
        model.add(GRU(50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
        model.add(GRU(50))
        model.add(Dense(1))
        model.compile(optimizer=Adam(), loss='mean_squared_error')
        model.fit(x_train, y_train, epochs=10, batch_size=1, verbose=0)

        # Predict
        predictions = model.predict(x_test)
        predictions_rescaled = scaler.inverse_transform(predictions)
        y_test_rescaled = scaler.inverse_transform(y_test)

        # Calculate metrics
        mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y_test_rescaled, predictions_rescaled)
        r2 = r2_score(y_test_rescaled, predictions_rescaled)
        accuracy_percent = r2 * 100

        # Store metrics
        mse_list.append(mse)
        rmse_list.append(rmse)
        mae_list.append(mae)
        r2_list.append(r2)
        accuracy_list.append(accuracy_percent)

        # Print individual metrics
        print(f"Metrics for {filename}:")
        print(f"  MSE: {mse:.4f}")
        print(f"  RMSE: {rmse:.4f}")
        print(f"  MAE: {mae:.4f}")
        print(f"  R¬≤ Score: {r2:.4f}")
        print(f"  Accuracy (%): {accuracy_percent:.2f}%")

        # Add prediction column (optional)
        full_df['Prediction'] = np.nan
        full_df.iloc[-len(predictions):, full_df.columns.get_loc('Prediction')] = predictions_rescaled.flatten()

        # Save predictions to CSV ---
         output_path = os.path.join(output_folder, filename.replace('.csv', '_with_prediction.csv'))
         full_df.to_csv(output_path, index=False)
         print(f"‚úÖ Saved predictions to: {output_path}")

        # Plot
        plt.figure(figsize=(12, 6))
        plt.plot(df['Date'], df['Close'], label='Actual')
        plt.plot(df.iloc[-len(predictions):]['Date'], predictions_rescaled, label='Predicted')
        plt.title(f'{filename} - Stock Price Prediction (GRU)')
        plt.xlabel('Date')
        plt.ylabel('Stock Price')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"‚ùå Error processing {filename}: {e}")

# Print average metrics
if mse_list:
    print("\nüìä Average Metrics Across All Stocks:")
    print(f"  Average MSE       : {np.mean(mse_list):.4f}")
    print(f"  Average RMSE      : {np.mean(rmse_list):.4f}")
    print(f"  Average MAE       : {np.mean(mae_list):.4f}")
    print(f"  Average R¬≤ Score  : {np.mean(r2_list):.4f}")
    average_accuracy = np.mean(accuracy_list)
    print(f"  Average Accuracy (%): {average_accuracy:.2f}%")
else:
    print("No metrics calculated.")
```
